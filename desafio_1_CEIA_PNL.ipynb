```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
texto = "The dogs are running and studies were done on better methods." #probamos

tokens = word_tokenize(texto.lower())
tokens_alpha = [t for t in tokens if t.isalpha()]

print("Tokens:", tokens_alpha)
```

```python
stems = [stemmer.stem(t) for t in tokens_alpha]
print("Stemming:", stems)
```

```python
lemmas = [lemmatizer.lemmatize(t) for t in tokens_alpha]
print("Lemmatization:", lemmas)
```

Siguiendo el pipeline de NLP clásico presentado en la clase, se incorporó una etapa de lemmatization antes de la vectorización. A diferencia del stemming, que puede producir raíces truncadas, la lemmatization devuelve palabras válidas y preserva mejor la semántica, aunque requiere recursos lingüísticos adicionales. Por esta razón se eligió lemmatization para preprocesar los textos antes de aplicar TF-IDF. Seguimos con el paso 5 que es la vectorización.

```python
def lemmatize_text(text):#funcion de lematizacion para todo el corpus
    tokens = word_tokenize(text.lower())#pasamos a minusculas / separa el texto en palabras
    tokens_alpha = [t for t in tokens if t.isalpha()] #elimina puntuacion numeros y simbolos
    lemmas = [lemmatizer.lemmatize(t) for t in tokens_alpha]#convierte palabras a su forma base
    return " ".join(lemmas)
```

```python
train_docs_lem = [lemmatize_text(doc) for doc in newsgroups_train.data]# lo aplicamos a los documentos de train
test_docs_lem = [lemmatize_text(doc) for doc in newsgroups_test.data]
```

2. Construir un modelo de clasificación por prototipos (tipo zero-shot).

    Clasificar los documentos de un conjunto de test comparando cada uno con todos los de entrenamiento y asignar la clase al label del documento del conjunto de entrenamiento con mayor similaridad.


```python
#cada documento de test comparado contra todos los de train
similarity_test_train = cosine_similarity(X_test_lem, X_train_lem)
print(similarity_test_train.shape)
```

```python
#para cada documento test devuelve, el doc de train más parecido
nearest_train_idx = np.argmax(similarity_test_train, axis=1)
print(nearest_train_idx[:10])
```

```python
#le asigno la clase del documento más similar
y_test = newsgroups_test.target
y_pred_proto = y_train[nearest_train_idx]
```

```python
results = []

vectorizers = {
    "tfidf_base": TfidfVectorizer(stop_words='english', max_df=0.7, min_df=2),
    "tfidf_strict": TfidfVectorizer(stop_words='english', max_df=0.5, min_df=5),
    "count": CountVectorizer(stop_words='english', max_df=0.7, min_df=2)
}

alphas = [0.1, 0.5, 1.0]

for vect_name, vect in vectorizers.items():
    
    X_train = vect.fit_transform(train_docs_lem)
    X_test = vect.transform(test_docs_lem)

    for alpha in alphas:
        
        # Multinomial
        model_m = MultinomialNB(alpha=alpha)
        model_m.fit(X_train, y_train)
        pred_m = model_m.predict(X_test)
        f1_m = f1_score(y_test, pred_m, average='macro')

        results.append((vect_name, "MultinomialNB", alpha, f1_m))

        # Complement
        model_c = ComplementNB(alpha=alpha)
        model_c.fit(X_train, y_train)
        pred_c = model_c.predict(X_test)
        f1_c = f1_score(y_test, pred_c, average='macro')

        results.append((vect_name, "ComplementNB", alpha, f1_c))
```

```python
results_sorted = sorted(results, key=lambda x: x[3], reverse=True)

for r in results_sorted:
    print(r)
```

Se probaron distintos modelos de Naïve Bayes cambiando tanto el vectorizador como algunos hiperparámetros. El mejor resultado se obtuvo usando TF-IDF con la configuración base junto con ComplementNB y α = 0.5, logrando un F1-score macro de 0.678. En general, TF-IDF funcionó mejor que CountVectorizer, lo que muestra que es importante ponderar las palabras según su relevancia en el corpus. Además, ComplementNB superó a MultinomialNB, algo esperable ya que maneja mejor el desbalance entre clases. Por último, ajustar el valor de α permitió mejorar un poco el rendimiento, siendo un valor intermedio el que dio mejores resultados.

4. Transponer la matriz documento-término.

    De esa manera se obtiene una matriz término-documento que puede ser interpretada como una colección de vectorización de palabras.
    Estudiar ahora similaridad entre palabras tomando 5 palabras y estudiando sus 5 más similares.

Elegir las palabras MANUALMENTE para evitar la aparición de términos poco interpretables.

```python
# Paso 1: transponer matriz documento-término
X_terms = X_train_lem.T
print("Shape término-documento:", X_terms.shape)

# Paso 2: diccionarios de palabras
word2idx = tfidfvect_lem.vocabulary_
idx2word = {v: k for k, v in word2idx.items()}

# Paso 3: elegir palabras manualmente
selected_words = ["car", "space", "god", "game", "window"]

# Verificar que existan
for w in selected_words:
    print(w, w in word2idx)

# Paso 4: obtener índices
selected_indices = [word2idx[w] for w in selected_words if w in word2idx]

# Paso 5: calcular similitud
similarity_words = cosine_similarity(X_terms[selected_indices], X_terms)
print("Shape matriz de similaridad:", similarity_words.shape)

# Paso 6: mostrar top 5 similares
for i, idx in enumerate(selected_indices):
    sims = similarity_words[i]
    most_sim = np.argsort(sims)[::-1][1:6]

    print("\n" + "="*80)
    print(f"Palabra base: {idx2word[idx]}")
    print("Top 5 palabras más similares:")
    
    for j in most_sim:
        print(f"{idx2word[j]} - Similaridad: {sims[j]:.4f}")
```

Los resultados muestran que la similitud coseno aplicada sobre la matriz término-documento permite capturar relaciones temáticas entre palabras. Las palabras más similares no son necesariamente sinónimos, sino términos que aparecen en contextos similares dentro del corpus. Esto evidencia que el modelo aprende patrones de co-ocurrencia, agrupando palabras según su uso en diferentes dominios (automotriz-car, espacial-space, religioso-god, deportivo-game, informático-window).
